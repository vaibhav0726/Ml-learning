Matplotlib Title

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0, 1, 2, 3])
ypoints = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(xpoints, ypoints)
plt.title("sales")
xpoints = np.array([0, 1, 2, 3])
ypoints = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(xpoints, ypoints)
plt.title("income")
plt.suptitle("My Shop")
plt.show()
```
